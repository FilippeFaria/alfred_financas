# Análise do Fluxo de Caixa

Este notebook permite explorar e analisar os dados do fluxo de caixa.

```python
import pandas as pd
import plotly.express as px

# Carregar dados do fluxo de caixa
# O arquivo usa ponto e vírgula como delimitador e encoding ISO-8859-1
df = pd.read_csv(
    'fluxo_de_caixa.csv',
    sep=';',
    encoding='latin-1',
    parse_dates=['Data'],
    dayfirst=True
)

# Exibir informações básicas
print(f"Total de registros: {len(df)}")
print(f"\nColunas disponíveis:")
print(df.columns.tolist())
print(f"\nPrimeiros registros:")
df.head(10)
```

```python
# Resumo por tipo (Receita/Despesa/Investimento)
print("Resumo por Tipo:")
print(df['Tipo'].value_counts())

print("\n\nResumo por Categoria:")
print(df['Categoria'].value_counts())
```

```python
# Converter coluna Valor para numérico (substituir vírgula por ponto)
df['Valor_Num'] = df['Valor'].astype(str).str.replace(',', '.').astype(float)

# Filtrar registros não desconsiderados
df_filtrado = df[df['desconsiderar'] == False].copy()

# Resumo por ano/mês
df_filtrado['AnoMes'] = df_filtrado['Data'].dt.to_period('M')

resumo_mensal = df_filtrado.groupby(['AnoMes', 'Tipo'])['Valor_Num'].sum().unstack(fill_value=0)
resumo_mensal['Saldo'] = resumo_mensal.sum(axis=1)
print("Resumo Mensal:")
resumo_mensal
```

```python
# Gráfico de evolução do saldo mensal
resumo_mensal_plot = resumo_mensal.reset_index()
resumo_mensal_plot['AnoMes'] = resumo_mensal_plot['AnoMes'].astype(str)

fig = px.bar(
    resumo_mensal_plot, 
    x='AnoMes', 
    y='Saldo',
    title='Saldo Mensal ao Longo do Tempo',
    labels={'AnoMes': 'Mês', 'Saldo': 'Saldo (R$)'}
)
fig.show()
```

```python
# Despesas por categoria (gráfico)
despesas = df_filtrado[df_filtrado['Tipo'] == 'Despesa']
categoria_despesa = despesas.groupby('Categoria')['Valor_Num'].sum().sort_values()

fig = px.bar(
    categoria_despesa.reset_index(),
    x='Valor_Num',
    y='Categoria',
    orientation='h',
    title='Despesas por Categoria',
    labels={'Valor_Num': 'Total (R$)', 'Categoria': 'Categoria'}
)
fig.show()
```
